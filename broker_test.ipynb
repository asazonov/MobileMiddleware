```python
import pandas as pd
import matplotlib as mpl
import glob
import numpy
import random
```

```python
data_list = []
file_list = []
file_list = glob.glob("./output/*.csv")
for file in file_list:
    tmp_data = pd.read_csv(file)
    data_list.append(tmp_data)
```

```python
data_list
```

```python
plt.figure(figsize=(30,30))
for data in data_list:
    data.plot(y="time", style = "x", c=numpy.random.rand(3,))
#plt.savefig("broker-test.png", format='png')
```
